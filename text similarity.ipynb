# Text similarity between two sentences using spacy

```python
import pandas as pd
```

```python
# importing spacy
import spacy
```

```python
# reading the csv file
data = pd.read_csv('Precily_Text_Similarity.csv')
# printing the shaps of the csv file
print(data.shape)
data
```

```python
# loading en_core_web_md pipeline
nlp = spacy.load('en_core_web_md')
```

```python
# spacy tokenizes text and produces doc object
doc1 = [nlp(sentence) for sentence in data['text1']]
```

```python
doc1[0]
```

```python
# spacy tokenizes text and produces doc object
doc2 = [nlp(sentence) for sentence in data['text2']]
```

```python
# similarity method from spacy is used to check the similarity between two sentences
similarity = []
for i in range(len(doc1)):
    similarity.append(doc1[i].similarity(doc2[i]))
```

```python
similarity
```

```python
demo1 = nlp('i am going out')
demo2 = nlp('i am leaving')
```

```python
demo1.similarity(demo2)
```

```python
demo1 = nlp('i am going out')
demo2 = nlp('i am going out')
demo1.similarity(demo2)
```

```python
data['similarity']= similarity
```
